Escribir un programa que pregunte al usuario por las ventas de un rango de años y muestre por pantalla una serie con los datos de las ventas indexada por los años, antes y después de aplicarles un descuento del 10%.

```python
import pandas as pd
# 62, 2341, 4564, 2354, 7453, 5235, 45453, 53643, 26421, 326425, 23421

# Lo voy a hacer con funciones.

def get_input():
  # Lo primero, inicialmente, sería pedir los datos, claramente.
  ventas = input('Introduzca sus ventas entre el año 2012 y 2022 separadas por comas (",")     ---------->  ').split(',')
  ano = 2012

  prices = pd.DataFrame() # DataFrame de Pandas vacio.

  for venta in ventas:
    venta = float(venta)
    prices[ano] = [venta]
    ano += 1
  print(f'SIN DESCUENTO:\n{prices}')
  ano = 2012

  for decuentos in range(len(ventas)):
    prices.loc[0, ano] -= (prices.loc[0, ano] * 10) / 100
    ano += 1



  print(f'\n\nCON DESCUENTO DEL 10%:\n{ prices }')






get_input();
```

```python
# Para separar un DataFrame por grupos es bastante sencillo, y para ello, utilizamos
# DF.groupby([columnas]).groups; De esta manera, se separará y nos devolverá un diccionario 
# con una lista, que dentro lleva el valor (o valores) de las diferentes columnas seleccionadas,
# etiquetadas por esos valores.

import pandas as pd
import numpy as np



pacientes = pd.read_csv('https://raw.githubusercontent.com/asalber/manual-python/master/datos/colesterol.csv')
print(pacientes.groupby(['sexo']).groups, end='\n\n\n')



# Podríamos tambien obtener un grupo en concreto, para esto usamos DF.groupby(['sexo']).get_group(grupos). Es importante
# que la lista y la tupla de grupos tengan el mismo tamaño.
print(pacientes.groupby(['sexo']).get_group('M'), end='\n\n\n')





# Una método bastante util para mostrar y organizar los grupos y a la vez darle 
# una función de agregación a ese grupo, es usar DF.groupby([columnas]).agg(funcion_de_agregacion).
# Recordemos que una función de agrecion es aquella que recibe una lista de valores, pero solamente devuelve 1.
# Algunos ejemplos puede ser: La media, el maximo o el minimo.

print(pacientes.groupby(['sexo']).agg(np.std), end='\n\n\n')
```

```python
# Vamos a mirar como reestructurar un DF de formato largo a formato largo. Esto nos puede servir para distintas situaciones,
# bien sean a la inversa, o como ya nombramos anteriormente, vamos a necesitar casi seguramente en algún momento
# construir de una forma distinta los datos para que nos sea más sencillo trabajar.

# Para reestructurar del formato largo al formato largo, usamos:
# DF.melt(id_vars=id-columnas, value_vars=columnas, var_name=nombre-columnas, var_value=nombre-valores)

import pandas as pd

datos = {'nombre':['María', 'Luis', 'Carmen'],
          'edad':[18, 22, 20],
          'Matemáticas':[8.5, 7, 3.5],
          'Economía':[8, 6.5, 5],
          'Programación':[6.5, 4, 9]}


datos_df = pd.DataFrame(datos)

datos_df_largo = datos_df.melt(id_vars=['nombre', 'edad'], var_name='asignatura', value_name='nota') # El id_vars no se moverá. 
print(datos_df_largo, end='\n\n\n')



# Para hacer la operación inversa, haremos uso de DF.pivot(index=filas, columns=columna, values=[valores]). 
# Con esto, se crearán n columnas nuevas, donde n es el numero de valores unicos que haya en columna. 
# Con esto me refiero a que si hay Matematicas y Economia 3 veces en "columna", solo se creará una columna por cada valor.
print(datos_df_largo.pivot(index='nombre', columns='asignatura', values='nota'), end='\n\n\n')
```

```python
# Cuando estemos trabajando con distintos DataFrames, pero que tengan relación entre sí,
# nos va a ser imprescindible unir estos datos para trabajar de forma comoda, es por ello
# que con pandas podemos unir 2 DataFrames diferentes: ya se a por la concatenación de filas,
# o ya sea por la concatenación de columnas.

# Cuando queremos concatenar por filas, es importantisimo que los dos DataFrame's tengan
# el mismo indice de columnas, mientras que cuando queramos concatenar por columnas, las 
# filas tienen que tener los mismos indices.


# Para concatenar cualqueir cantidad de DataFrames, usamos DF.concat([list_of_dataframe's], axis = 1 o 0)
# Algo que tener en cuanta cuando queremos concatenar los DataFrames, son el eje. El eje toma como valores 1 o 0.
# Si axis = 0, como se concatenarán, será por filas, mientras que si es 1, será por columnas.




# Concatenación por filas.

import pandas as pd

DB1 = pd.DataFrame({'Name': ['Fernando', 'Milena', 'Cristian', 'Luis', 'Maria'], 'Sex': ['H', 'M', 'H', 'H', 'M'], 'Age': [16, 42, 19, 52, 10]})
DB2 = pd.DataFrame({'Name': ['Jose Luis', 'Crespo', 'Xiomara', 'Ari', 'Juan'], 'Sex': ['H', 'H', 'M', 'M', 'H'], 'Age': [23, 62, 8, 25, 30]})

BD_D = pd.concat([DB1, DB2]) # Si lo dejamos sin "axis", por defecto el valor de él es 0

print(BD_D, end='\n\n\n')






# concatenación por columnas:
DB_01 = pd.DataFrame({'Name': ['Fernando', 'Milena', 'Cristian', 'Luis', 'Maria'], 'Sex': ['H', 'M', 'H', 'H', 'M']}).set_index('Name') # Importante establecer el indice principal.
DB_02 = pd.DataFrame({'Name': ['Fernando', 'Milena', 'Cristian', 'Luis', 'Maria'], 'Age': [23, 62, 8, 25, 30]}).set_index('Name')  # Importante establecer el indice principal.

BD_F = pd.concat([DB_01, DB_02], axis = 1)
print(BD_F, end='\n\n\n')







# MEZCLAR DATAFRAMES.
# Mezclar DataFrames nos permite juntar los mismos que tienen información en común, pero que hay columnas
# que un DF las tiene y otro no.
# Para hacer este tipo de "mezclas", hacemos uso del método "pd.merge(DF1, DF2, on = clave, how = tipo)"

# Clave guarda una lista de columnas "clave", que son las columnas en común. how es el tipo de mezcla que se utiliza
# (ya parece esto panadería XD). How puede tomar diferentes valores, por defecto es "inner", que el DF final
# es uno que solo contienen los valores de "on" que se encuentran en los  2 DataFrame.
# "outer" es otra propiedad recibida por "how", esta sirve para mostrar absolutamente todas las filas de "on",
# haciendo que, si no está en la otra lista, los valores que recogerá serán de tipo desconocido (NaN).
# También tenemos "left", que lo que hace es que deja todos las filas de primer DataFrame (sin excepción),
# y concatena las columnas no estén en ese DF y que coincidan con el "on", en el caso de que ninguna de las claves
# se pueda concatenar, simplemente no se mete. Luego también tenemos 'right', que es lo mismo, pero del revés.
# Esto quiere decir que toma como referencia el segundo DataFrame y trata de concatenar ese, con el primero.


DB_01 = pd.DataFrame({'Name': ['Fernando', 'Milena', 'Cristian'], 'Sex': ['H', 'M', 'H']})
DB_02 = pd.DataFrame({'Name': [ 'Cristian', 'Fernando', 'Milena', 'Luisiano'], 'Age': [23, 62, 8, 25]})

DB_0F = pd.merge(DB_01, DB_02, on='Name', how='outer') # !!!!!!
print(DB_0F)
```

# ***Ejercicio 2***

> Escribir una función que reciba un diccionario con las notas de los alumno de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.







```python
import pandas as pd
import numpy as np

def processing_notes(notas_dict):
  students_notes = pd.Series(notas_dict)
  
  minima = students_notes.agg(np.min)
  maxima = students_notes.agg(np.max)
  media = students_notes.agg(np.mean)
  desviacion =  students_notes.agg(np.std)
  
  return pd.Series([minima, maxima, media, desviacion], index=['Valor Mínimo', 'Valor Máximo', 'Media', 'Desviacion Típica'])
  



notas = {'Juan':9, 'María':6.5, 'Pedro':4, 'Carmen': 8.5, 'Luis': 5}
processing_notes(notas)
```

# ***Ejercicio 3***

> Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con las notas de los alumnos aprobados ordenadas de mayor a menor.




```python
import pandas as pd


def alumnos_aprobados(notas_dict):
  alumnos = pd.Series(notas_dict).sort_values(ascending = True)
  return alumnos[alumnos > 5]







notas = {'Juan':9, 'María':6.5, 'Pedro':4, 'Carmen': 8.5, 'Luis': 5}

print(alumnos_aprobados(notas))
```

Construir una función que construya un DataFrame a partir del un fichero con el formato anterior y devuelva otro DataFrame con el mínimo, el máximo y la media de dada columna.

```python
import pandas as pd

import numpy as np



def toDF(filecsv):
  my_DataFrame = pd.read_csv(filecsv, sep=';', decimal=',')
  new_volume_fix = []

  for i in my_DataFrame['Volumen']:
    a = i
    a = a.replace('.', '')
    a = float(a)
    new_volume_fix.append(a)
    





  alist = [[np.min(my_DataFrame['Final']), np.max(my_DataFrame['Final']), np.mean(my_DataFrame['Final'])], [np.min(my_DataFrame['Máximo']), np.max(my_DataFrame['Máximo']), np.mean(my_DataFrame['Máximo'])], [np.min(my_DataFrame['Mínimo']), np.max(my_DataFrame['Mínimo']), np.mean(my_DataFrame['Mínimo'])], [my_DataFrame['Volumen'].agg(np.min), my_DataFrame['Volumen'].agg(np.max), np.mean(new_volume_fix)],[np.min(my_DataFrame['Efectivo']), np.max(my_DataFrame['Efectivo']), np.mean(my_DataFrame['Efectivo'])]]
  
  ND = pd.DataFrame(alist, index=['Final', 'Máximo', 'Mínimo', 'Volumen', 'Efectivo'], columns=['Mínimo', 'Máximo', 'Media'])
  print(ND)



  # print(my_DataFrame)

toDF('/content/cotizacion.csv')
```
