```python
# Ejercicio 9
# Escribir un programa que pregunte al usuario la fecha de su nacimiento en formato 
# dd/mm/aaaa y muestra por pantalla, el día, el mes y el año. Adaptar el programa 
# anterior para que también funcione cuando el día o el mes se introduzcan con un solo carácter.



datev = input('\n\n\tBuen día!\n\nIntroduce su fecha en formato dd/mm/aaaa -----------> ')
day = "0" + datev[:datev.find('/')]
lenday = len(day)
month = "0" + datev[lenday:datev.find('/', lenday)]
year = "0" + datev[-4:]
print(f' Día {day[-2:]} // Mes {month[-2:]} // Año {year[-4:]} ')
```

```python
# Ejercicio 10
# Escribir un programa que pregunte por consola por los productos de una cesta de 
# la compra, separados por comas, y muestre por pantalla cada uno de los productos 
# en una línea distinta.


# Método 1 (eficiente)



my_list = input('Tipee su lista del supermercado separadas por comas ----------->  ')
nl = my_list.split(',')

for i in nl:
  print(f'\n\t- {i}')




# método 2

new_list = my_list.replace(',', '\n')
print(new_list)
```
