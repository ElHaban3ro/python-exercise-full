```python
# Ejercicio 3
# Escribir un programa que pida al usuario un número entero positivo y muestre por 
# pantalla todos los números impares desde 1 hasta ese número separados por comas.

num_impares = int(input('Ingrese un numero entero positivo --------> '))
base_impares = 1
 
if num_impares % 2 == 0:
  num_impares -= 1

num_impares = round(num_impares / 2)

for i in range(num_impares):
  print(base_impares, end=', ')
  base_impares += 2
```

```python
# Ejercicio 5
# Escribir un programa que pregunte al usuario una cantidad a invertir, el interés 
# anual y el número de años, y muestre por pantalla el capital obtenido en la 
# inversión cada año que dura la inversión.

invertir = float(input('¿Cuanto quiere (dinero) invertir? ---------> '))
interes = float(input('¿Cuanto es el interés anual? ---------> '))
tiempo = int(input('¿Cuanto tiempo (años) quiere invertir? ---------> '))


for inversion in range(tiempo):
  invertir *= 1 + interes / 100

  print(f'Año {inversion + 1}: { round(invertir, 2) }')
```

```python
# Ejercicio 6
# Escribir un programa que pida al usuario un número entero y muestre por pantalla 
# un triángulo rectángulo como el de más abajo, de altura el número introducido.

altura = int(input('Para crear un triangulo necesito que me des la altura ------> '))

for base in range(altura):
  print('*' * (base + 1), end='\n')
```

```python
# Ejercicio 8
# Escribir un programa que pida al usuario un número entero y muestre por pantalla 
# un triángulo rectángulo como el de más abajo.


numT = int(input('Introduce un numero ---------- > '))


for i in range(1, numT + 1, 2):
  for j in range(i, 0, -2):
    print(j, end=' ')
  
  print('')
```

```python
# Ejercicio 10
# Escribir un programa que pida al usuario un número entero y muestre por pantalla 
# si es un número primo o no.

num_primo = int(input('Ingrese un número para saber si es primo -------> '))

for i in range(2, num_primo):
  if num_primo % i == 0:
    print('\n\n\t>> El numero ignresado no es primo.')
    break
  else:
    print('\n\n\t>> El numero ingresado es primo')
    break
```

```python
# Ejercicio 11
# Escribir un programa que pida al usuario una palabra y luego muestre por pantalla 
# una a una las letras de la palabra introducida empezando por la última.

word = input('Ingrese su palabra para invertirla --------> ')[::-1]

for i in word:
  print(i, end=' ')
```

```python
# Ejercicio 12
# Escribir un programa en el que se pregunte al usuario por una frase y una letra, 
# y muestre por pantalla el número de veces que aparece la letra en la frase.

phrase = input('Ingrese una frase!  - - - - - - - - - - - - >   ')
letter = input('Ingrese una letra  - - - - - - - - - - - - >   ')
coincidencias = 0

for l in phrase:
  if letter == l:
    coincidencias += 1
    print(f'\n\n\t>> La letra {letter} aparece {coincidencias} veces...')
```

```python
# Ejercicio 13

# Escribir un programa que muestre el eco de todo lo que el usuario introduzca hasta 
# que el usuario escriba “salir” que terminará.


notas = input(' Notas >   ')
print(f'\n\t{notas}')

while notas != 'salir':
  notas = input('\n\n\n Notas >   ')
  print(f'\n\t{notas}')
  if notas == 'salir':
    continue

print('\n\n\t---- Ya no es "hola mundo", ahora es "adios mundo" :(( ----')  
```
