```python
# Ejercicio 1
# Escribir una función que pida un número entero entre 1 y 10 y guarde en un fichero 
# con el nombre tabla-n.txt la tabla de multiplicar de ese número, done n es el 
# número introducido.




def save_number():
  my_number = int(input('Introduce un numero entre 1 y 10. ------------->   '))

  open_file = open(f'tabla-{my_number}.txt', 'a')


  for tm in range(1, 11):

    open_file.write(f'{my_number} x {tm} = {my_number * tm}\n')

  open_file.close()




save_number()
```

```python
# Ejercicio 5
# Escribir un programa que abra el fichero con información sobre el PIB per cápita 
# de los países de la Unión Europea 
# (url:https://ec.europa.eu/eurostat/estat-navtree-portlet-prod/BulkDownloadListing?file=data/sdg_08_10.tsv.gz&unzip=true), 
# pregunte por las iniciales de un país y muestre el PIB per cápita de ese país de todos los años disponibles.

import requests


def per_capita_europa():
  url = 'https://ec.europa.eu/eurostat/estat-navtree-portlet-prod/BulkDownloadListing?file=data/sdg_08_10.tsv.gz&unzip=true'
  pc = dict()

  get = requests.get(url)
  read = get.text.split('\n')

  for rl in read[1:]:
    rl = list(rl.split('\t'))
    pais = rl[0].split(',')

    pc[pais[-1]] = rl[1:]


  years = read[0].split('\t')

  opcion = input('¿Qué país quieres ver? ----------------->    ').upper()


  if opcion in pc:
    print('AÑO\tPIB\n')
    for p in range(1, len(pc[opcion]) + 1):
      print(f'{years[p]}\t{pc[opcion][p - 1]}')
    








    





per_capita_europa()
```

```python
# Ejercicio 6
# Escribir un programa para gestionar un listín telefónico con los nombres y los 
# teléfonos de los clientes de una empresa. El programa incorporar funciones crear 
# el fichero con el listín si no existe, para consultar el teléfono de un cliente, 
# añadir el teléfono de un nuevo cliente y eliminar el teléfono de un cliente. El 
# listín debe estar guardado en el fichero de texto listin.txt donde el nombre del 
# cliente y su teléfono deben aparecer separados por comas y cada cliente en una 
# línea distinta.


def listar_clientes():
  
  f = open('listin.txt', 'a+')
  f.close()


  f = open('listin.txt', 'r')

  list_of_lines = f.readlines() # Devuelve una lista.

  
  if len(list_of_lines) > 0:
    for lines in list_of_lines:
      print(f' - {lines}')


  else:
    print('\n\n\t\tSu libreta de contactos está vacia!')
  
  f.close()





def anadir_clientes():

  nombre = input('\n\n\t\tIntroduzca el nombre del contacto a continuación ------------>   ')
  numero = input('\t\tIntroduzca el numero del contacto a continuación ------------>   ')

  f = open('listin.txt', 'a+')
  escritura = f.write(f'{nombre}, {numero}\n')
  print('\n\t\t\tHemos añadido un contacto nuevo!')
  f.close()



def eliminar_cliente():
  numero = input('\n\n\t\tIntroduzca el numero del contacto a eliminar ------------>   ')
  deleted = False


  f = open('listin.txt', 'a+')
  f.close()

  f = open('listin.txt', 'r+')
  list_of_contacts = f.readlines()

  
  if len(list_of_contacts) > 0:

    for n in list_of_contacts:

      list_split = n.split(',')
      list_split[1] = list_split[1].replace(' ', '')
      list_split[1] = list_split[1].replace('\n', '')
      # print(list_spli6t)

      if numero == list_split[1]:
        print(f'\n\n\t\tHemos eliminado el usuario con telefono {numero}.')
        list_of_contacts.remove(n)
        f.truncate(0)
        f.seek(0)
        f.writelines(list_of_contacts)
        deleted = True


    if deleted != True:  
      print(f'\t\tEl numero {numero} ingresado no coincide con ninguno de la lista\n\n')




  else:
    print('\n\n\t\tSu libreta de contactos está vacia!')
  





def home(opcion):
  if opcion == 1:
    return listar_clientes()

  elif opcion == 2:
    return anadir_clientes()

  elif opcion == 3:
    return eliminar_cliente()





opcion = 0

while opcion != 4:
  
  opcion = int(input('\n\t\t------ Agenda de contactos -------\n\n¿Qué quiere hacer?\n\t\t\t(1) Consultar telefonos\n\t\t\t(2) Añadir Contacto\n\t\t\t(3) Eliminar contacto\n\t\t\t(4) Salir\n\t\t\t-------------->   '))
  home(opcion)

```

```python
from numpy import nan
# Ejercicio 8
# El fichero calificaciones.csv contiene las calificaciones de un curso. Durante 
# el curso se realizaron dos exámenes parciales de teoría y un examen de prácticas. 
# Los alumnos que tuvieron menos de 4 en alguno de estos exámenes pudieron repetirlo 
# en la al final del curso (convocatoria ordinaria). Escribir un programa que contenga 
# las siguientes funciones:

# (1) Una función que reciba el fichero de calificaciones y devuelva una lista de 
# diccionarios, donde cada diccionario contiene la información de los exámenes y la 
# asistencia de un alumno. La lista tiene que estar ordenada por apellidos.

# (2) Una función que reciba una lista de diccionarios como la que devuelve la función 
# anterior y añada a cada diccionario un nuevo par con la nota final del curso. El 
# peso de cada parcial de teoría en la nota final es de un 30% mientras que el peso 
# del examen de prácticas es de un 40%.

# (3) Una función que reciba una lista de diccionarios como la que devuelve la función
# anterior y devuelva dos listas, una con los alumnos aprobados y otra con los 
# alumnos suspensos. Para aprobar el curso, la asistencia tiene que ser mayor o 
# igual que el 75%, la nota de los exámenes parciales y de prácticas mayor o igual 
# que 4 y la nota final mayor o igual que 5.

import pandas as pd
import math
import numpy as np

# Función 1.

def to_dict_func(file):
  f = open(file, 'a+') # Es importante abrirlo ya que parece que se buguea si no se hace.
  f.close()
  apellidos = []
  

  a_dict = pd.read_csv(file, sep=';').to_dict('records')

  for claves in a_dict: # Obtemenos los apellidos.
    apellido = claves['Apellidos']
    apellidos.append(apellido) # Los meto en una lista.
  apellidos = sorted(apellidos) # Ya en una lista, es más sencillo ordenarlos por orden alfabetico.

  list_ordenada = list() # Creamos una nueva lista, la que ya estará ordenada.
  for last in apellidos: # Recorremos apellidos, que usaremos como validación.

    for elem_list in range(len(a_dict)): # Para cada apellido, recorremos todo el diccionario.
      if last == a_dict[elem_list]['Apellidos']: # Vemos si el diccionario actual en el campo "Apellidos"...
      # coincide con nuestro apellido actual (ya ordenado).
        list_ordenada.append(a_dict[elem_list]) # Si coincide, metemos a la nueva lista ese elemento que coincide


  return list_ordenada



def nota_final(list_values):
  for dicts in enumerate(list_values):
    notas = []
    notas = dicts[1]['Parcial1'], dicts[1]['Parcial2'], dicts[1]['Ordinario1'], dicts[1]['Ordinario2'], dicts[1]['Practicas'], dicts[1]['OrdinarioPracticas']
    # print(notas)
    Parciales = list(notas[:4])
    Practicas = list(notas[4:])

    for par_values in Parciales:
      if pd.isna(par_values) == True:
        Parciales[Parciales.index(par_values)] = 0
        
      else:
        try:
          number_converted = par_values.replace(',', '.')
          Parciales[Parciales.index(par_values)] = float(number_converted)

        except AttributeError:
          pass



    for pra_values in Practicas:
      if pd.isna(pra_values) == True:
        Practicas[Practicas.index(pra_values)] = 0

      else:
        try:
          number_converted_dos = pra_values.replace(',', '.')
          Practicas[Practicas.index(pra_values)] = float(number_converted_dos)

        except AttributeError:
          pass



    Parciales = round((sum(Parciales) * 30) / 70, 1)
    Practicas = round((sum(Practicas) * 40) / 70, 1)

    nota_final = Parciales + Practicas

    dicts[1]['Nota Final Parciales'] = Parciales
    dicts[1]['Nota Final Practicas'] = Practicas
    dicts[1]['Nota Final'] = round(nota_final / 2, 3)



  return list_values




def aprobados(estudiantes_list):
  aprobados = []
  reprobados = []

  for estudiantes in estudiantes_list:
    if float(estudiantes['Asistencia'].replace('%', '')) >= 75 and estudiantes['Nota Final Parciales'] >= 4 and estudiantes['Nota Final Practicas'] >= 4 and estudiantes['Nota Final'] >= 5.0:
      aprobados.append(estudiantes)

    else:
      reprobados.append(estudiantes)

  print(f'Alumnos Aprobados: {aprobados}\n\n\n\nAlumnos Reprobados: {reprobados}')
  return 






mi_lista = to_dict_func('calificaciones.csv')

notas_finales = nota_final(mi_lista)


print(aprobados(notas_finales))
```
