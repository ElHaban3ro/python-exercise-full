```python
# Ejercicio 3
# Escribir una función que reciba un número entero positivo y devuelva su factorial



def factorial(numero):
  factorial = 1
  for factorial_de in range(numero):
    if factorial_de >= 1:
      factorial *= factorial_de
  return factorial * (factorial_de + 1)



# n = int(input('Introduzca un numero'))

print(factorial(4))
print(factorial(20))
```

```python
# Ejercicio 6
# Escribir una función que reciba una muestra de números en una lista y devuelva 
# su media.

def media(list_of_numbers):
  media_of_numbers = 0
  for number in list_of_numbers:
    media_of_numbers = (media_of_numbers + number)

  return media_of_numbers  / len(list_of_numbers)




media([1, 2, 3, 4, 5])
```

```python
# Ejercicio 7
# Escribir una función que reciba una muestra de números en una lista y devuelva 
# otra lista con sus cuadrados.

def cuadrados(list_of_numbers2):
  cuadrados = []
  for number2 in list_of_numbers2:
    number2 = number2 **2
    cuadrados.append(number2)
  return cuadrados




print(cuadrados([1, 2, 3, 4, 5]))
print(cuadrados([2.3, 5.7, 6.8, 9.7, 12.1, 15.6]))
```

```python
# Ejercicio 8
# Escribir una función que reciba una muestra de números en una lista y devuelva 
# un diccionario con su media, varianza y desviación típica.

import numpy as np



def statics(list_of_numbers3):

  estadisticas = dict()

  # Media
  media_of_numbers = 0
  for number in list_of_numbers3:
    media_of_numbers = (media_of_numbers + number)
  media_of_numbers  /= len(list_of_numbers3)

  estadisticas['Media'] = media_of_numbers


  # Varianza
  varianza = np.var(list_of_numbers3)
  estadisticas['Varianza'] = varianza


  # Desviación típica.
  desviacion = np.std(list_of_numbers3)
  estadisticas['Desviación Típica'] = desviacion






  return estadisticas






print(statics([1, 2, 3, 4, 5]))
```

```python
# Ejercicio 9
# Escribir una función que calcule el máximo común divisor de dos números y otra 
# que calcule el mínimo común múltiplo.
import math

def maximo_comun_divisor(n1, n2):
  print(f'Maximo común divisor: {math.gcd(n1, n2)}')
  return 

def minimo_comun_divisor(n1, n2):
  
  # Para sacar el lcm es importante primero saber cual es el numero más grande.
  if n1 > n2:
    mayor = n1
  else:
    mayor = n2

  
  while True:
    if mayor % n1 == 0 and mayor % n2 == 0:
      lcm = mayor
      break
    mayor += 1

  return 'El minimo  común multiplicador es:', lcm



maximo_comun_divisor(24,36)
minimo_comun_divisor(24,36)
```

```python
# Ejercicio 10
# Escribir una función que convierta un número decimal en binario y otra que convierta 
# un número binario en decimal.
import math


def compute_binary(n1):

  binarios = ''
  while n1 / 2 != 0:

    if n1 % 2 == 1:
      n1 = n1 / 2
      n1 = math.floor(n1)
      binarios += '1'
      # print(1)

    elif n1 % 2 == 0:
      n1 = n1 / 2
      n1 = math.floor(n1)
      binarios += '0'
      # print(0)

    elif n1 == 1:
      break

  print(binarios[::-1])

compute_binary(23)




def compute_decimal(n):
  n = str(n)[::-1]
  binary_converted = 0    
  exponente = 0
  for binario in n:
      binary_converted += int(binario) * (2**exponente)
      exponente += 1

  print(binary_converted)
  return




compute_decimal(10111)
```
