```python
# Para empezar a dibujar con matplotlib es importante primero importar la librera.
import matplotlib.pyplot as plt
import numpy as np

# Las principales funciones de matplotlib son:
# figure(num, figsize, dpi, facecolor, edgecolor, frameon)
  # Ac, num, sirve para ennumerar la figura. Esto lo podemos dejar en falso para que este procedimiento se haga solo.
  # figsize es el tamao del grafico. Este recibe una tupla de (ancho, alto), en pulgadas.
  # Seguidamente de figsize es importante ver "dpi", usada para establecer una resolucin indicada para el grafico.
  # El facecolor sirver para establecer el color de fondo del grafico.
  # Y, con edgecolor, establecemos el color del marco de la figura, donde vienen ennumerados los datos.
  # Y, frameon establecido en "False", nos quita el marco de datos. Util en determinados momentos.


# subplot(numRows, numCols, plotNum)
  # Con subplot podemos meter varias graficas en un solo marco.
  # Ac, numRows estabelce el numero de filas, y respectivamente, numCols establece el numero de columnas.
  # plotNum establece el numero de la grafica.


# Luego, usamos plot(x, y, linestyle, linewidth, marker) para un grafico con varios graficos dentro.
  # x, bsicamente representa el eje de las abcisas.
  # y, respectivamente representa las ordenadas. x e y pueden ser bien tuplas o listas, pero es importante que las 2 tengan el mismo tamao
  # con linestyle podemos establecer el estilo de la linea, tanto el color, como tambin cmo est dibujada. Puede ser con puntos, continua, punteada...
      # Algunos de los tipos de lineas comunes son:
        #  '-' Solida, 
        # '-.' Linea con puntos y rayas, 
        # '--' Linea cortada 
        # o ':', que es linea punteada.
        
      # El tipo de linea va a compaada antes por el color de la linea, y se usa las iniciales de los colores para pintarlas.
        # 'c' Cin.
        # 'b' Azul.
        # 'g' Verde.
        # 'y' Amarillo.
        # 'k' Negro.
        # 'w' Blanco.
        # 'r' Rojo.
        # 'm' Magenta.
        # ...




a = np.linspace(0, 20, 100)
b = np.sin(a)


plt.plot(a, b, 'r-', linewidth = 3)
plt.show()
```

# ***Ejercicio 1***
> Escribir un programa que pregunte al usuario por las ventas de un rango de aos y muestre por pantalla un diagrama de lneas con la evolucin de las ventas.

```python
import matplotlib.pyplot as plt

# 3156, 6431, 3446, 5413, 7462, 8468, 6458, 7456, 10879, 11847

def graphic_sales():
  inicio = input('Introduzca el rango de ao en el que empienzan sus ventas ------->  ')
  ventas = input(f'Introduzca sus ventas desde {inicio} hasta 2022 separadas por coma  ------------>   ').split(',')
  ventas_float = []

  for venta in ventas:
    venta = float(venta)
    ventas_float.append(venta)


  ventas_float.sort()

  anos_list = []
  for anos in range(int(inicio), 2023):
    anos_list.append(anos)





  plt.figure(linewidth = 2, figsize = (10, 10))
  plt.plot(anos_list, ventas_float, 'r-o')
  plt.xlabel('AOS DE VENTAS')
  plt.ylabel('VENTAS')
  plt.title('MIS VENTAS!')
  

graphic_sales()
```

# *Ejercicio 2*
> Escribir una funcin que reciba una diccionario con las notas de las asignaturas de un curso y una cadena con el nombre de un color y devuelva un diagrama de barras de las notas en el color dado.

```python
import matplotlib.pyplot as plt

def print_students_notes(dict_notes, color_bar):
  colors = {'Rojo': 'r', 'Verde': 'g', 'Amarillo': 'y', 'Negro': 'k', 'Azul': 'b'}

  if color_bar in colors:
    asignatura = list(dict_notes.keys())
    nota = list(dict_notes.values())

    plt.bar(asignatura, nota, align = 'center', color = colors[color_bar])
    plt.title('NOTAS PROMEDIO UNIVERSIDAD')
    plt.xlabel('ASIGNATURAS')
    plt.ylabel('NOTAS (0 - 10)')
    plt.show()
    

  return



notas_a = {'Programacin':9, 'Mates':6.5, 'Economa':4, 'Historia': 8}




print(print_students_notes(notas_a, 'Rojo'))
```

```python
import pandas as pd
import matplotlib.pyplot as plt


def diagram_to_png(series, nombre_de_archivo):
  fig = plt.figure(figsize = (3, 3), dpi = 300)
  plt.title('VENTAS', c = '#960020')
  plt.pie(series, labels = series.index, autopct="%0.1f%%", colors = ['#C70039', '#900C3F', '#581845'])
  plt.show
  plt.savefig(f'{nombre_de_archivo}.png')







ventas_tienda = {'Enero':200, 'Febrero':240, 'Marzo':310}
diagram_to_png(pd.Series(ventas_tienda), 'mis_ventas_primeros_meses_2022')
```

# ***Ejercicio 5***
> Escribir una funcin que reciba una serie de Pandas con el nmero de ventas de un producto por aos y una cadena con el tipo de grfico a generar (lineas, barras, sectores, areas) y devuelva un diagrama del tipo indicado con la evolucin de las ventas por aos y con el ttulo Evolucin del nmero de ventas.

```python
import pandas as pd
import matplotlib.pyplot as plt


def your_option(datos, grafico_opcion):
  def grafico_barras():
    fig = plt.figure(figsize = (5, 5))
    plt.bar(datos.index, datos)
    plt.title('EVOLUCIN DEL NMERO DE VENTAS', fontsize = 15)
    plt.show()


  def grafico_lineas():
    fig = plt.figure(figsize = (10, 5))
    plt.plot(datos.index, datos, '-o')
    plt.title('EVOLUCIN DEL NMERO DE VENTAS', fontsize = 15)
    plt.show()


  def grafico_pastel():
    
    fig = plt.figure(figsize = (5, 5))
    plt.pie(datos, labels=datos.index, autopct="%0.1f%%")
    plt.title('EVOLUCIN DEL NMERO DE VENTAS', fontsize = 15)
    plt.show()


  def grafico_area():

    fig = plt.figure(figsize = (10, 10))

    plt.fill_between(datos.index, datos)

    plt.title('EVOLUCIN DEL NMERO DE VENTAS', fontsize = 15)
    plt.show()







  opciones = {'Barras': grafico_barras, 'Lineas': grafico_lineas, 'Pastel': grafico_pastel, 'Area': grafico_area}
  if grafico_opcion in opciones:
    opciones[grafico_opcion]()




df_ventas = pd.Series([1200, 840, 1325, 1280, 1500], index = [2000, 2001, 2002, 2003, 2004])
your_option(df_ventas, 'Barras')
your_option(df_ventas, 'Lineas')
your_option(df_ventas, 'Pastel')
your_option(df_ventas, 'Area')

```

# ***Ejercicio 7***
> El fichero bancos.csv contiene las cotizaciones de los principales bancos de Espaa con : Empresa (nombre de la empresa), Apertura (precio de la accin a la apertura de bolsa), Mximo (precio mximo de la accin durante la jornada), Mnimo (precio mnimo de la accin durante la jornada), Cierre (precio de la accin al cierre de bolsa), Volumen (volumen al cierre de bolsa). Construir una funcin reciba el fichero bancos.csv y cree un diagrama de lneas con las series temporales de las cotizaciones de cierre de cada banco.

```python
import pandas as pd
import matplotlib.pyplot as plt


def files_to_graphics(file):

  # Leer archivo csv.
  read_file = pd.read_csv(file)
  
  # Eliminando columnas que no nos interesan:
  read_file = read_file.drop(['Apertura', 'Mximo', 'Mnimo', 'Volumen'], axis = 1)
  fig = plt.figure(figsize = (40, 10), dpi = 200)



  grupos = read_file.groupby(['Empresa'])
  
  for banco in grupos.groups:
    plt.plot(grupos.get_group(banco)['Fecha'], grupos.get_group(banco)['Cierre'], '-o')
  
  plt.ylabel('Precio de Cierre')
  plt.ylim([0.2, 6])

  plt.xlabel('Fechas')
  plt.legend(['BBVA', 'Banco Sabadell', 'Banco Santander', 'Bankinter', 'Caixabank'], loc = 'lower left')
  plt.title('COTIZACIONES', fontsize = 40)
  plt.show()




files_to_graphics('bancos.csv')
```
