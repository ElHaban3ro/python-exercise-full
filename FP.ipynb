```python
def decoratoras(function):
  def func_return(*args):  
    print('Función de retorno.')
    retorno = function(*args)
    return retorno
  
  return func_return




@decoratoras
def saludo(nombre):
  print(f'Hello {nombre}')


saludo('Fernando')
```

```python
# Filter

def pares(numbers):
  if numbers % 2 == 0:
    return True


list_num = [2, 45, 13, 4, 9, 1, 8]

print(list(filter(pares, list_num)))
```

```python
class Empleado:
  def __init__(self, nombre, cargo, salario):
    self.nombre = nombre
    self.cargo = cargo
    self.salario = salario

  def __str__(self):
    return f'{self.nombre}, que trabaja como {self.cargo} ahora tiene un salario de {self.salario}€'



lista_empleados = [
                   
                   Empleado("Fernando", "Desarrollador", 150),
                   Empleado("Wos", "Él", 1000000),
                   Empleado("Juan José", "Marketing", 25000),
                   Empleado("Sandra Milena", "CEO", 1500000)

]


def calculo(empleado):

  if empleado.salario <= 30000:
    empleado.salario = empleado.salario * 1.10
    empleado.salario = round(empleado.salario)

  return empleado


lista_empleados = map(calculo, lista_empleados)

for em in lista_empleados:
  print(em)
```
