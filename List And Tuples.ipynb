```python
# Ejercicio 4
# Escribir un programa que pregunte al usuario los números ganadores de la lotería 
# primitiva, los almacene en una lista y los muestre por pantalla ordenados de menor a mayor.


loteria = input('Introduzca los numeros ganadores de la lotería ------------->   ').split(',')
lotint = []

for l in loteria:
  converted = int(loteria[loteria.index(l)])
  lotint.append(converted)



print(f'\n\n\n-------------- Ganadores de la lotería --------------') # No se cuenta, es simple decoración

lotint = sorted(lotint)
for winner in lotint:
  print(f'\n\t>> {winner}')
```

```python
from numpy.lib.function_base import median
# Ejercicio 13
# Escribir un programa que pregunte por una muestra de números, separados por comas, 
# los guarde en una lista y muestre por pantalla su media y desviación típica.

import numpy as np

numbers = input('Proporcione números separados por comas para sacar su media y su desviación tipica -------> ').split(',')
nn = []
promedio = 0

for n in numbers:
  conversion = float(n)
  nn.append(conversion)


# Media 
for media in nn:
  promedio += media
promedio /= len(nn)


print(np.std(nn))
print(promedio)

```
