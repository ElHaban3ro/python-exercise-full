```python
# Ejercicio 7
# Escribir un programa que cree un diccionario simulando una cesta de la compra. 
# El programa debe preguntar el artículo y su precio y añadir el par al diccionario, 
# hasta que el usuario decida terminar. Después se debe mostrar por pantalla la 
# lista de la compra y el coste total, con el siguiente formato


cesta = {}
total = 0

print('--------------- Bienvenido a la cesta de la compra ---------------\n')

add_product_name = input('\n\n\n\n\t> Añadir producto --------------- > ').capitalize()
add_product_price = float(input('\n\n\t> ¿Cual es el precio del producto? --------------- > '))
cesta[add_product_name] = add_product_price
# Bucle While

while add_product_name != 'Salir':
  


  add_product_name = input('\n\n\n\n\t> Añadir producto --------------- > ').capitalize()
  add_product_price = float(input('\n\n\t> ¿Cual es el precio del producto? --------------- > '))
  cesta[add_product_name] = add_product_price


else:
  cesta_products = list(cesta.keys())
  cesta_descriptions = list(cesta.values())
  for product in range(len(cesta) - 1):
    print(f'\n\t\t> {cesta_products[product]}\t | \t{cesta_descriptions[product]}$')
    total += cesta_descriptions[product]
  
  print(f'\n\t\t> TOTAL \t | \t {total}$')


```

```python
# Ejercicio 8
# Escribir un programa que cree un diccionario de traducción español-inglés. El 
# usuario introducirá las palabras en español e inglés separadas por dos puntos, y 
# cada par <palabra>:<traducción> separados por comas. El programa debe crear un 
# diccionario con las palabras y sus traducciones. Después pedirá una frase en español y 
# utilizará el diccionario para traducirla palabra a palabra. Si una palabra no 
# está en el diccionario debe dejarla sin traducir.



dict_ing_virgen = {}
dict_ing = input('Introduzca una frase, donde cada palabra debe llevar el formato palabra:traduccion, y cada par de estos, separados por coma\n\n ------------>   ').lower().replace(' ', '').split(',')


# Preparar el diccionario.
for palabra in range(len(dict_ing)):
  separacion = dict_ing[palabra].split(':')

  dict_ing_virgen[separacion[0]] = separacion[1]


# Separacion del diccionario en listas para poder acceder más facil.
espanol = list(dict_ing_virgen.keys())
ingles = list(dict_ing_virgen.values())

# print(dict_ing_virgen)

# # Hacer traduccion.
frase = input('Ahora introduzca cualquier frase y la traduciremos -----------------  >   ').lower()
frase_complete = frase
frase = frase.split(' ')
print(frase_complete)

for traduccion in frase:

  for idk in espanol:
    if frase[frase.index(traduccion)] == espanol[espanol.index(idk)]:
      frase_complete = frase_complete.replace(f' {traduccion}', f' {ingles[espanol.index(traduccion)]}')

print(f'\n\n\t> traducción: \n\n\t\t {frase_complete}')




# me:I, gusta:like, papa:potato, pc:pc, carro:car, auto:car, hola:hello, cómo:how, con:with, y:and, yo:I, tu:you, jugar: play
# Hola soy Fernando y me gusta jugar juegos de pc aunque mas que todo los de auto
```

```python
# Ejercicio 9
# Escribir un programa que gestione las facturas pendientes de cobro de una empresa. 
# Las facturas se almacenarán en un diccionario donde la clave de cada factura será 
# el número de factura y el valor el coste de la factura. El programa debe preguntar 
# al usuario si quiere añadir una nueva factura, pagar una existente o terminar. 
# Si desea añadir una nueva factura se preguntará por el número de factura y su coste y 
# se añadirá al diccionario. Si se desea pagar una factura se preguntará por el 
# número de factura y se eliminará del diccionario. Después de cada operación el
# programa debe mostrar por pantalla la cantidad cobrada hasta el momento y la cantidad 
# pendiente de cobro.

facturas = {}
opciones = 0
mostrar = 0
asd = 0
pagadas = 0

print('\n--------------------------------------------------\n| \t\t ADMINISTRACION \t\t | \n--------------------------------------------------\n\n')


while opciones != 4:
  print('-------------------------------------------------------------------------------------------------------------------------------------------')
  opciones = int(input('\n\n\n\n¿Qué quiere hacer? \n\n\t\t> (1) Añadir factura!\n\t\t> (2) Pagar factura existente!\n\t\t> (3) Listar facturas!\n\t\t> (4) Salir\n\t\t\t | > |  '))



  if opciones == 1:
    print('\n\n\n\n\t\t> Añade una factura!')

    add_id_invoice = input('\n\t\t\t Añada el numero de factura >  ')
    add_price_invoice = float(input('\n\t\t\t Añada el precio de la factura >  '))
    facturas[add_id_invoice] = add_price_invoice

    facturas_id_list = list(facturas.keys())
    facturas_valor_list = list(facturas.values())
    sin_pagar_list = sum(facturas_valor_list)


    print(f'\n\n\t\t\t\t > Pentiende por cobrar:  {sin_pagar_list}\t\t|\t > Cobrado:  {pagadas}')
    continue


  elif opciones == 2:
    print('\n\n\n\n\t\t> Eliminar una factura!')
    delete_id_invoice = input('\n\t\t\t Indica el numero de factura para eliminar >  ')


    if delete_id_invoice not in facturas_id_list:
      print('\n\n------------------------------------------ ⚠ ¡ASEGURESE DE QUE EL NUMERO DFACTURA ES CORRECTO! ⚠ ------------------------------------------ \n\n')
      continue
    
    else:
      sin_pagar_list -= facturas[delete_id_invoice]
      pagadas += facturas[delete_id_invoice]
      del facturas[delete_id_invoice]

      print(f'\n\n\t\t\t\t > Pendiente por cobrar:  {sin_pagar_list}\t\t|\t > Cobrado:  {pagadas}')

      continue



  elif opciones == 3:
    if len(facturas) == 0:
      print('\n\n------------------------------------------ ⚠ ¡NO PODEMOS LISTAR FACTURAS QUE NO EXISTEN jaja! ⚠ ------------------------------------------\n\n')
    else:
      print('\n\n\n\n================================ FACTURAS ================================\n|\t\tN. Factura\t\t\t\tPrecio\t')
      for listar_id in range(len(facturas)):
          print(f'| \t\t {facturas_id_list[listar_id]} \t\t\t\t\t {facturas_valor_list[listar_id]}\t\t |')

    continue




  else:
    print('\n\nAdiós!')
    break




  # print(f'\n\n\t\t\t\t > Pentiende por cobrar:  {sin_pagar}$\t\t ------- \t > Cobrado:  {pagadas}$')
```

```python
# Ejercicio 10
# Escribir un programa que permita gestionar la base de datos de clientes de una 
# empresa. Los clientes se guardarán en un diccionario en el que la clave de cada 
# cliente será su NIF, y el valor será otro diccionario con los datos del cliente 
# (nombre, dirección, teléfono, correo, preferente), donde preferente tendrá el 
# valor True si se trata de un cliente preferente. El programa debe preguntar al 
# usuario por una opción del siguiente menú: (1) Añadir cliente, (2) Eliminar cliente, 
# (3) Mostrar cliente, (4) Listar todos los clientes, (5) Listar clientes preferentes, 
# (6) Terminar. En función de la opción elegida el programa tendrá que hacer lo siguiente:

  # Preguntar los datos del cliente, crear un diccionario con los datos y añadirlo a la base de datos.
  # Preguntar por el NIF del cliente y eliminar sus datos de la base de datos.
  # Preguntar por el NIF del cliente y mostrar sus datos.
  # Mostrar lista de todos los clientes de la base datos con su NIF y nombre.
  # Mostrar la lista de clientes preferentes de la base de datos con su NIF y nombre.
  # Terminar el programa.

import time # Libreria que tiene varios métodos para distintas cosas de tiempo que necesitemos. En este caso usaré time.sleep()



# Creamos un diccionario vacio desde acá.
DB = dict()

# Es importante instanciar esta variable para que entre directamente al bucle.
opcion = 0

print('\n------------------------------------------\n| \t\t DB USERS \t\t | \n------------------------------------------\n\n')



print()
while opcion != 6:
  # Preguntamos la opcion sobre escribiendo el valor de opcion.
  opcion = int(input('\n\n\t> (1) Añadir cliente\n\t> (2) Eliminar cliente\n\t> (3) Mostrar cliente\n\t> (4) Listar todos los clientes\n\t> (5) Listar clientes preferentes\n\t> (6) Terminar\n\n\t\t¿Qué quiere hacer? ---------->   '))


  if opcion == 1:

    # Obtención de datos.
    print('\n\n\tUste va a añadir un nuevo usuario con los siguientes datos:  ')
    add_user_NIF = input(f'\t\t NIF -------------->  ')
    add_user_nombre = input(f'\n\n\t\t Nombre Completo -------------->  ').capitalize()
    add_user_direccion = input(f'\n\n\t\t Dirección -------------->  ')
    add_user_telefono = int(input(f'\n\n\t\t Telefono -------------->  '))
    add_user_email = input(f'\n\n\t\t Correo Electronico -------------->  ')
    add_user_superuser = (input(f'\n\n\t\t ¿Es superusuario? (solo usuarios autorizados) -------------->  '))


    # "PUT" de los datos. Los guardamos el diccionario "DB". recibe en el valor de "add_user_NIF" otro diccionario.
    DB[add_user_NIF] = {'Nombre': add_user_nombre, 'Direccion': add_user_direccion, 'Telefono': add_user_telefono, 'Email': add_user_email, 'SuperUser': add_user_superuser}
    print(f'\n\t\t\t>>>Hemos añadido el usario con NIF {add_user_NIF}\n\n\n\n')
    time.sleep(5)



  elif opcion == 2:
    
    # Preguntamos por el NIF para eliminar esa clave del diccionario.
    print('\n\n\tUste va a eliminar a un usuario, proporcione el NIF: ')
    delete_user_NIF = input(f'\t\t NIF -------------->  ')
    
    if delete_user_NIF in DB:
      del DB[delete_user_NIF]
      print(f'\nHemos eliminado el usario con NIF {delete_user_NIF}\n\n\n\n')
      time.sleep(5)

    else:
      print('\n\n\t\t\t>>> Asegurese de que ese NIF exista!\n\n\n\n')
      time.sleep(5)




  elif opcion == 3:
    # Preguntamos por el NIF para eliminar esa clave del diccionario.
    print('\n\n\tUste va a consultar la información de un usuario, proporcione el NIF: ')
    consult_user_NIF = input(f'\t\t NIF -------------->  ')

    if consult_user_NIF in DB:
      print(f'\n| ================================== User {consult_user_NIF} ================================== |\n\n\tNombre\t\t\t\t\t{ DB[consult_user_NIF]["Nombre"] } \n\n\tEmail\t\t\t\t\t{ DB[consult_user_NIF]["Email"] } \n\n\tTelefono\t\t\t\t\t{ DB[consult_user_NIF]["Telefono"] } \n\n\tDireccion\t\t\t\t\t{ DB[consult_user_NIF]["Direccion"] } \n\n\tisSuperUser\t\t\t\t\t{ DB[consult_user_NIF]["SuperUser"] }\n\n| ================================== User {consult_user_NIF} ================================== |\n\n\n\n\n123')
      time.sleep(5)



  elif opcion == 4:
    print('\n\n\tUste va a consultar la información de todos los usuarios...\n\n')

    if len(DB) >= 1:
      print('\n| ================================== Users DB ================================== |')
      print(f'\t\t NOMBRE\t\t NIF\t\t isSuperUser132\n\n')
      for user in DB:
        print(f'\n\t\t{DB[user]["Nombre"]}\t\t{user}\t\t{DB[user]["SuperUser"]}\n\n')
      time.sleep(5)


      print('\n| ================================== Users DB ================================== |')


    else:
      print('\n\n\t\t\t>>> No hay usuarios para mostrar!\n\n\n\n')      
      time.sleep(5)



  elif opcion == 5:
    print('\n\n\tUste va a consultar los super usuarios...\n\n')
    superUsers_count = 0
    superUsers_list = []
    superUsers_NIF_list = []
    for superUsers in DB:
      if DB[superUsers]['SuperUser'] == 'True':
        superUsers_list.append(DB[superUsers])
        superUsers_NIF_list.append(superUsers)  
        superUsers_count += 1

    print(superUsers_NIF_list)
    time.sleep(5)



    if superUsers_count >= 1:
      print('\n| ================================== SUPERUSERS DB ================================== |')
      print(f'\t\t NOMBRE\t\t NIF\n\n')
      for superUser_only in superUsers_list:
        print(f'\n\t\t{ superUser_only["Nombre"] }\t\t{ superUsers_NIF_list[superUsers_list.index(superUser_only)] }\n\n')

      print('\n| ================================== SUPERUSERS DB ================================== |')
      time.sleep(5)
    

    else:
      print('\n\n\t\t\t>>> No hay SUPER USUARIOS PARA MOSTRAR!\n\n\n\n')  
    time.sleep(5)
```

```python
# Ejercicio 11
# El directorio de los clientes de una empresa está organizado en una cadena de 
# texto como la de más abajo, donde cada línea contiene la información del nombre, 
# email, teléfono, nif, y el descuento que se le aplica. Las líneas se separan con 
# el carácter de cambio de línea \n y la primera línea contiene los nombres de los 
# campos con la información contenida en el directorio.
# "nif;nombre;email;teléfono;descuento\n01234567L;Luis González;luisgonzalez@mail.com;656343576;12.5\n71476342J;Macarena Ramírez;macarena@mail.com;692839321;8\n63823376M;Juan José Martínez;juanjo@mail.com;664888233;5.2\n98376547F;Carmen Sánchez;carmen@mail.com;667677855;15.7"

# Escribir un programa que genere un diccionario con la información del directorio, donde cada elemento corresponda a un cliente y tenga por clave su nif y por valor otro diccionario con el resto de la información del cliente. Los diccionarios con la información de cada cliente tendrán como claves los nombres de los campos y como valores la información de cada cliente correspondientes a los campos. Es decir, un diccionario como el siguiente.
# {'01234567L': {'nombre': 'Luis González', 'email': 'luisgonzalez@mail.com', 'teléfono': '656343576', 'descuento': 12.5}, '71476342J': {'nombre': 'Macarena Ramírez', 'email': 'macarena@mail.com', 'teléfono': '692839321', 'descuento': 8.0}, '63823376M': {'nombre': 'Juan José Martínez', 'email': 'juanjo@mail.com', 'teléfono': '664888233', 'descuento': 5.2}, '98376547F': {'nombre': 'Carmen Sánchez', 'email': 'carmen@mail.com', 'teléfono': '667677855', 'descuento': 15.7}}




info_virgen = "nif;nombre;email;teléfono;descuento\n01234567L;Luis González;luisgonzalez@mail.com;656343576;12.5\n71476342J;Macarena Ramírez;macarena@mail.com;692839321;8\n63823376M;Juan José Martínez;juanjo@mail.com;664888233;5.2\n98376547F;Carmen Sánchez;carmen@mail.com;667677855;15.7".split('\n')

descuentos_clientes = dict()

descuentos =  dict()


# Palabras clave para usarlas como claves en el diccionario
key_words = info_virgen[0].split(';')


# Datos (claves:valores)
for datos in info_virgen[1:]: # Recorremos la información virgen para separar los 
# datos, esto lo hacemos quitando el indice del NIF, ya que no queremos que se muestre.
# Lo utilizaremos como Key.

  info = info_virgen[info_virgen.index(datos)].split(';') # Separamos la información ahora sí.

  for i in range(len(info_virgen[1:])): # Creamos un for en el rango de la información empezando desde 1 (sin el NIF).
    descuentos[key_words[i + 1]] = info[i +1]


  descuentos_clientes[info[0]] = descuentos


print(descuentos_clientes)
```
